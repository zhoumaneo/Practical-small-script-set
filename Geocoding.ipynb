```python
import pandas as pd
import numpy as np
import requests
import json
```

```python
airport = pd.read_csv("airport.csv",encoding = "GBK")
```

```python
airport
```

```python
headers = {
    'user-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.97 Safari/537.36'
}
```

```python
airport[['lng','lat']] = pd.DataFrame([None,None]).T
lng = []
lat = []
```

```python
for i in range(len(airport)):
    api_url = "http://api.map.baidu.com/geocoding/v3/?address="+str(airport.iloc[i]['name'])+"&output=json&ak=BTAOo85Vv3ArOvy5FGUzG1Yy5CkrG0eG"
    jsondata = json.loads(requests.get(api_url,headers=headers).text)
    lng.append(jsondata['result']['location']['lng'])
    lat.append(jsondata['result']['location']['lat'])
airport[['lng','lat']] = pd.DataFrame([lng,lat]).T
airport.to_csv('output.csv')
```
