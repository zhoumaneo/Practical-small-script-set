```python
from urllib import request,parse
```

```python
data.columns
```

```python
import json
```

```python
req_url = "http://fanyi.youdao.com/translate"
From_Date = {}
From_Date['i'] = 'I love you'
From_Date['doctype'] = 'json'
text = parse.urlencode(From_Date).encode('utf-8')
response = request.urlopen(req_url,text)
html = response.read().decode('utf-8')
translate_result = json.loads(html)
translate_result = translate_result['translateResult'][0][0]['tgt']
translate_result
```

```python
Proposal_Title = []
for i in range(len(data)):
    From_Date['i'] = data.iloc[i]['Proposal Title']
    text = parse.urlencode(From_Date).encode('utf-8')
    response = request.urlopen(req_url,text)
    html = response.read().decode('utf-8')
    translate_result = json.loads(html)
    translate_result = translate_result['translateResult'][0][0]['tgt']
    Proposal_Title.append(translate_result)
```
